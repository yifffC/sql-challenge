```python
import pandas as pd
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:password@localhost:5432/EmployeeSQL')
connection = engine.connect()
```

```python
salary_data = pd.read_sql("SELECT * FROM salaries", connection)
salary_df = pd.DataFrame(salary_data)
salary_df
```

```python
title_data = pd.read_sql("SELECT * FROM titles", connection)
title_df = pd.DataFrame(title_data)
title_df
```

```python
employee_data = pd.read_sql("SELECT * FROM employees", connection)
employee_df = pd.DataFrame(employee_data)
employee_df
```

```python
#Create a histogram to visualize the most common salary ranges for employees.
salary_df.hist(column='salary', bins=10)
```

```python
#Create a bar chart of average salary by title
emp_salary_df = salary_df.merge(employee_df, on='emp_no')
emp_salary_title_df = emp_salary_df.merge(title_df, left_on='emp_title_id', right_on='title_id')
salary_title_df = emp_salary_title_df.drop(columns=['emp_no', 'emp_title_id', 'birth_date', 'first_name', 'last_name', 'sex', 'hire_date', 'title_id'])
average_salary_df = round(salary_title_df.groupby(['title']).mean(), 2)
average_salary_df
```
